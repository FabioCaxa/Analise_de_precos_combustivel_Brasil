## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
# Carregando os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combinando-os em um único DataFrame.
preco2000 = pd.read_csv("gasolina_2000+.csv", index_col=0)
preco2010 = pd.read_csv("gasolina_2010+.csv", index_col=0)
precos = pd.concat([preco2000, preco2010])
df = pd.DataFrame(precos)
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
# Convertendo as colunas "DATA INICIAL" e "DATA FINAL" para o tipo Datetime.
df["DATA INICIAL"] = pd.to_datetime(df["DATA INICIAL"])
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
type(df["DATA FINAL"].iloc[0])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
# Criando uma nova coluna representando o ano e o mês (aaaa-mm), utilizando a coluna DATA FINAL como referência.
df["ANO/MES"] = df["DATA FINAL"].dt.strftime("%Y-%m")
df
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
# Filtrando o DataFrame para obter apenas dados da 'GASOLINA COMUM' e gravando em uma nova variável.
gas_comum = pd.DataFrame(df[df["PRODUTO"] == "GASOLINA COMUM"])
gas_comum
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
# Calculando o preço médio de revenda da gasolina em agosto de 2008. Resolvi usar o módulo datetime para diferenciar a abordagem.
import datetime
gas2008 = gas_comum[(gas_comum['DATA FINAL'] > datetime.datetime(2008, 8, 1)) & (gas_comum['DATA FINAL'] < datetime.datetime(2008, 9, 1))]

pmr = gas2008["PREÇO MÉDIO REVENDA"].mean()

pmr
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
# Calculando o preço médio de revenda da gasolina em maio de 2014 em São Paulo.
gas_comum[(gas_comum["ANO/MES"] == "2014-05") & (gas_comum["ESTADO"] == "SAO PAULO")]["PREÇO MÉDIO REVENDA"].mean()
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
# Identificando o(s) estado(s) onde a gasolina ultrapassou a barreira dos R$ 5,00 e quando isso ocorreu.
gas_comum[gas_comum["PREÇO MÉDIO REVENDA"] > 5][["ESTADO", "ANO/MES", "PREÇO MÉDIO REVENDA"]]
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
# Calculando a média de preço dos estados da região sul em 2012.
media_preco_2012_sul = gas_comum[(gas_comum["REGIÃO"] == "SUL") & (gas_comum["ANO/MES"].str.startswith("2012"))]["PREÇO MÉDIO REVENDA"].mean()
media_preco_2012_sul
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
# Obtendo uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro.
gas_comum['MES'] = gas_comum['DATA FINAL'].dt.month
rio_dados = gas_comum[gas_comum['ESTADO'] == 'RIO DE JANEIRO']
rio_ultimo_mes_ano = rio_dados.groupby(rio_dados['DATA FINAL'].dt.to_period('M'))[['PREÇO MÉDIO REVENDA', 'MES']].last()
rio_ultimo_mes_ano = rio_ultimo_mes_ano[rio_ultimo_mes_ano['MES'] == 12]
rio_pct_anual_corrigido = (rio_ultimo_mes_ano['PREÇO MÉDIO REVENDA'].pct_change() * 100)

rio_pct_anual_corrigido
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
# Criando uma tabela contendo uma série temporal com a diferença absoluta e percentual entre os valores mais baratos e caros.
preco_minimo_maximo = gas_comum.groupby('ANO/MES')['PREÇO MÉDIO REVENDA'].agg(['min', 'max']).reset_index()
preco_minimo_maximo
```

```python
preco_minimo_maximo['Diferença Absoluta'] = preco_minimo_maximo['max'] - preco_minimo_maximo['min']
preco_minimo_maximo
```

```python
preco_minimo_maximo["Diferença Percentual"] = (preco_minimo_maximo["Diferença Absoluta"] / preco_minimo_maximo["min"]) * 100
preco_minimo_maximo
```

```python
preco_minimo_maximo["ESTADO MAXIMO"] = gas_comum.loc[gas_comum.groupby('ANO/MES')['PREÇO MÉDIO REVENDA'].idxmax()]['ESTADO'].values
preco_minimo_maximo["ESTADO MINIMO"] = gas_comum.loc[gas_comum.groupby('ANO/MES')['PREÇO MÉDIO REVENDA'].idxmin()]['ESTADO'].values
```
